```python
x=[1,2,3,4,5]
y1=[5,1,7,8,2]
y2=[7,8,3,5,1]

plt.plot(x,y1,color="red",marker="o",linestyle="--",linewidth=3,markersize=6,label="line y1 as tata motors")
plt.plot(x,y2,color="b",marker="o",linestyle="--",linewidth=3,markersize=6,label="line y2 as tata power")

plt.title("multiple line in one plot")
plt.xlabel("Days")
plt.ylabel("stock price")
plt.legend()

plt.show()
```

```python
#pie chart>>To understand the percent distrubution of data
langs=["c","python","java"]
students=[20,100,40]

plt.pie(students,labels=langs,autopct="%1.1f%%")
plt.title("pie chart")
plt.show()
```

```python
#focus on one segment
explode=(0.0,0.1,0.1)
langs=["c","python","java"]
students=[20,100,40]

plt.pie(students,labels=langs,autopct="%1.1f%%",explode=explode)
plt.title("pie chart")
plt.show()
```

```python
explode=(0.0,0.1,0.1)
langs=["c","python","java"]
students=[20,100,40]

plt.pie(students,labels=langs,autopct="%1.1f%%",explode=explode,shadow=True)
plt.title("pie chart")
plt.show()
```

```python
#area plot
years=[2021,2018,2022,2023,2017]
category1=[20,25,30,28,35]
category2=[10,15,20,18,25]
category3=[5,8,10,9,15]

plt.stackplot(years,category1,category2,category3,labels=["category1","category2","category3"],colors=["skyblue","lightgreen","orange"])
plt.xlabel("years")
plt.ylabel("values")
plt.title("Area plot")
plt.legend()
plt.show()
```
