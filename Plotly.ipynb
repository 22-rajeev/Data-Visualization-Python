```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#graph_object
#express
import plotly.graph_objects as go
import plotly.express as px
```

```python
#using graph object
#add_trace>>adding data
#scatter plot
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[1,7,8,2,6],mode='markers'))
```

```python
#lines chart
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[1,7,8,2,6],mode='lines'))
```

```python
#using plotly express px
x=[1,2,3,4]
y=[10,13,15,18]
fig=px.scatter(x,y)
fig.show()
```

```python
fig=px.line(x,y)
fig.show()
```

```python
x=['A','B','C']
y=[10,5,6]
fig=px.bar(x=x,y=y)
fig.show()
```

```python
x=[5,10,3,4,5,5,6,10,3,11,11,12,3,4]
fig=px.histogram(x)
fig.show()
```

```python
x=['A','B','C']
y=[5,10,3,25,12,33,100]
fig=px.box(x,y)
fig.show()
```

```python
#plotting heatmap
data=[[1,2,3],[4,5,6],[7,8,9]]
fig=px.imshow(data)
fig.show()
```

```python
#bubble chart
x=[1,2,3,4]
y=[5,8,12,3]
fig=px.scatter(x,y,size=[30,10,50,100])
fig.show()
```

```python
#3d chart
x=[1,2,3,4]
y=[5,8,12,3]
z=[3,4,5,6]
fig=px.scatter_3d(x=x,y=y,z=z)
fig.show()
```

```python
#contour plot
x=[1,2,3,4,3,23,80]
y=[5,8,12,3,12,35,60]
fig=px.density_contour(x,y)
fig.show()
```
